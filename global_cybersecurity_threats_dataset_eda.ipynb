# üß† Exploratory Data Analysis: Global Cybersecurity Threats Dataset
This notebook performs a complete EDA process on the **Global Cybersecurity Threats Dataset** dataset.

```python
# üì• Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# üìä Setting styles
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12,6)

# üìÇ Load the dataset
df = pd.read_csv("Global_Cybersecurity_Threats_2015-2024.csv")

# üîç Preview the data
df.head()
```

```python
# üìã Basic Info
print("Shape of dataset:", df.shape)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())

# üìà Descriptive Statistics
df.describe(include='all')
```

```python
# üîç Univariate Analysis
df.hist(bins=20, figsize=(20, 15), color='skyblue')
plt.suptitle("Histograms of all Numerical Features", fontsize=20)
plt.show()
```

```python
# üìä Boxplots for detecting outliers
for column in df.select_dtypes(include=np.number).columns:
    plt.figure()
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.show()
```

```python
# üîó Correlation Heatmap (using only numeric columns)
numeric_df = df.select_dtypes(include=[np.number])  # select only numeric data
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features Only)")
plt.show()
```

```python
print("‚úÖ Dataset overview completed.\nNow based on the visuals and statistics above, we can draw insights regarding feature distribution, missing data, correlations, and potential outliers.")
```
